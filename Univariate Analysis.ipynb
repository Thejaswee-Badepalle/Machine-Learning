## Bivariate Analysis / Multivariate Analysis

#### 1. Does the cost of goods sold affect the ratings that the customers provide? (Numerical-Numerical)

```python
sns.scatterplot(x='cogs', y='Rating', data=df)
```

- There is no relationship b/w cost of goods sold(cogs) and rating

#### 2. Does gross income affect the ratings that the customers provide? (Numerical - Numerical)

```python
sns.scatterplot(x='gross income', y='Rating', data=df)
```

#### 3. Find the most profitable branch as per gross income(Numerical - Categorical)

```python
sns.barplot(x = 'Branch', y = 'gross income', data = df)
```

```python
# MultiVariate Analysis
sns.barplot(x = 'Branch', y = 'gross income', data = df, hue = 'City')
```

#### 4. Is there any relationship b/w gender and gross income (Numerical-Categorical)

```python
sns.boxplot(x = 'Gender', y = 'gross income', data = df)
```

```python
sns.boxplot(x = 'Gender', y = 'gross income', data = df, hue = 'Customer type')
```

#### 8. Which product line is purchased in the highest quantity?

```python
df.groupby('Product line').sum(numeric_only=True)['Quantity']
```

```python
(df.groupby('Product line').sum(numeric_only=True)['Quantity']).plot(kind = 'bar')
```

#### 9. Display daily sales by day of the week

```python
dw_mapping = {0:'Mon', 1:'Tue', 2:'Wed', 3:'Thur', 4: 'Fri', 5:'Sat', 6:'Sun'}
```

```python
df['day_of_week'] = df['Date'].dt.dayofweek.map(dw_mapping)
```

```python
df['day_of_week']
```

```python
df.head(1)
```

```python
df['day_of_week'].value_counts().plot(kind='bar')
```

#### 10. What will be the highest month for sales

```python
month_mapping = {1:"Jan", 2:"Feb", 3:"March"}
```

```python
df['month'] = df['Date'].dt.month.map(month_mapping)
```

```python
df.head(1)
```

```python
df['month'].value_counts().plot(kind = 'bar')
```
